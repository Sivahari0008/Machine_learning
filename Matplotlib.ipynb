```python
from mpl_toolkits import mplot3d
import numpy as np
```

```python
a = plt.figure()
b = a.add_subplot(111, projection='3d')

x = [randint(0,10) for i in range(3)]
y = [randint(0,10) for i in range(3)]
z = [randint(0,10) for i in range(3)]

b.plot3D(x,y,z,color='r',label="line")
plt.xlabel("x")
plt.ylabel("y")
b.set_zlabel("z")
plt.legend()



```

```python
a = plt.figure()
b = a.add_subplot(111, projection='3d')

x = [randint(0,10) for i in range(3)]
y = [randint(0,10) for i in range(3)]
z = [randint(0,10) for i in range(3)]

dx=np.ones(3)
dy=np.ones(3)
dz=[1,2,4]

b.bar3d(x,y,z,dx,dy,dz,color='r',label="line")
plt.xlabel("x")
plt.ylabel("y")
b.set_zlabel("z")
```
